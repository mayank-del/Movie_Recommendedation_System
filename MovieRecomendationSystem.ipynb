```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('/content/tmdb_5000_movies.csv')
credits=pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
# Here we will two dataSets by refrence of title column

movies=movies.merge(credits,on='title')
```

```python
'''
movies columns to be kept-->

#genres
#id
#keywords
#title
#overview
#cast
#crew


'''


movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
# since you can see overview column was having threee missing values
#so we are removing those 3 movies by dropna function so that we dont have null ovewviews
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
import ast
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#for cast
def convert3(obj):
  L=[]
  counter=0

  for i in ast.literal_eval(obj):
    if(counter!=3):
      L.append(i['name'])
    else:
      break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
# for crew
def fetch_director(obj):
  L=[]

  for i in ast.literal_eval(obj):
    if(i['job']=='Director'):
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python

movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
#text vectorisation
#in this technique every movies will be converted to vectos and we will catch movies of closest vectors
#as recomendataion


from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
#get all keywords formed in vector
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# we will not calcuate euclideam distance, it fails in higher dimesnion like dimension 5000
# hence we will cakculate cosine distance
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
 similarity
```

```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    #printing index
    print([i[0]])
    #priniting movie title
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
